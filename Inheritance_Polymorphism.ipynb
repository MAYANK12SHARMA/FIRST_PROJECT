Problem Statement: **Vehicle Hierarchy**

Create a Python program that models a hierarchy of vehicles using classes. Implement a superclass called Vehicle and two subclasses called Car and Motorcycle. Each class should have attributes and methods to represent specific characteristics of vehicles.

Class: Vehicle

Attributes:

make (string): Make or manufacturer of the vehicle.
model (string): Model name of the vehicle.
year (integer): Manufacturing year of the vehicle.

Methods:

display_info(self): Returns a string containing the details of the vehicle (year, make, model).

Class: Car (Subclass of Vehicle)

Additional Attributes:

num_doors (integer): Number of doors in the car.

Method Overrides:

display_info(self): Returns a string containing the details of the car, including the number of doors.

Class: Motorcycle (Subclass of Vehicle)

Additional Attributes:

num_wheels (integer): Number of wheels on the motorcycle.

Method Overrides:

display_info(self): Returns a string containing the details of the motorcycle, including the number of wheels.

Sample Input:

car1 = Car("Toyota", "Camry", 2022, 4)

motorcycle1 = Motorcycle("Harley-Davidson", "Street Glide", 2022, 2)

Sample Output:

print(car1.display_info())        # Output: 2022 Toyota Camry, 4-door car

print(motorcycle1.display_info())  # Output: 2022 Harley-Davidson Street Glide, 2-wheeler motorcycle

Explanation:

The program defines three classes: Vehicle, Car (subclass of Vehicle), and Motorcycle (subclass of Vehicle).

Each class has attributes to represent specific details of the vehicles, and the display_info method is overridden in the subclasses to provide additional details.

The sample input creates instances of Car and Motorcycle, and the sample output demonstrates calling the display_info method on each instance to obtain information about the vehicles.


```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make  #self._make = make
        self.model = model
        self.year = year

    def display_info(self):
        return f"{self.year} {self.make} {self.model}"

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def display_info(self):
        return f"{super().display_info()}, {self.num_doors}-door car"

class Motorcycle(Vehicle):
    def __init__(self, make, model, year, num_wheels):
        super().__init__(make, model, year)
        self.num_wheels = num_wheels

    def display_info(self):
        return f"{super().display_info()}, {self.num_wheels}-wheeler motorcycle"

# Creating object
car1 = Car("Toyota", "Camry", 2022, 4)
motorcycle1 = Motorcycle("Harley-Davidson", "Street Glide", 2022, 2)

print(car1.display_info())        # Output: 2022 Toyota Camry, 4-door car
print(motorcycle1.display_info())  # Output: 2022 Harley-Davidson Street Glide, 2-wheeler motorcycle
```

```python
#without using super()

class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        return f"{self.year} {self.make} {self.model}"

class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        # Explicitly calling the constructor of the parent class
        Vehicle.__init__(self, make, model, year)
        self.num_doors = num_doors

    def display_info(self):
        return f"{Vehicle.display_info(self)}, {self.num_doors}-door car"

class Motorcycle(Vehicle):
    def __init__(self, make, model, year, num_wheels):
        # Explicitly calling the constructor of the parent class
        Vehicle.__init__(self, make, model, year)
        self.num_wheels = num_wheels

    def display_info(self):
        return f"{Vehicle.display_info(self)}, {self.num_wheels}-wheeler motorcycle"

# Creating objects
car1 = Car("Toyota", "Camry", 2022, 4)
motorcycle1 = Motorcycle("Harley-Davidson", "Street Glide", 2022, 2)
```

Create a Python program that models an Animal Hierarchy using classes. Implement a superclass called Animal and two subclasses called Mammal and Bird. Each class should have attributes and methods to represent specific characteristics of animals.

**Class: Animal**

**Class: Mammal (Subclass of Animal)**

**Class: Bird (Subclass of Animal)**

Additional Attributes:

Method Overrides:

make_sound(self): Implements the sound a bird makes.

**Sample Input:**

**Sample Output:**

bird1.make_sound()          



```python
#Multiple inheritance without super()

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}")

class Employee:
    def __init__(self, emp_id, salary):
        self.emp_id = emp_id
        self.salary = salary

    def display_info(self):
        print(f"Employee ID: {self.emp_id}, Salary: ${self.salary}")

class Student(Person, Employee):
    def __init__(self, name, age, emp_id, salary, student_id, grade):
        # Call constructors of parent classes explicitly
        Person.__init__(self, name, age)
        Employee.__init__(self, emp_id, salary)
        self.student_id = student_id
        self.grade = grade

    def display_info(self):
        # Call display_info methods of parent classes explicitly
        Person.display_info(self)
        Employee.display_info(self)
        print(f"Student ID: {self.student_id}, Grade: {self.grade}")

# Usage
student = Student("John", 20, "E123", 50000, "S456", "A")
student.display_info()
```

# Polymorphism

**Problem Statement:**

Create a system to represent and display information about different geometric shapes using a class-based solution. Define classes for **Circle**, **Rectangle**, and **Triangle**, each with specific attributes and a method display_info to present details about the shape. Utilize polymorphism without using inheritance.

Classes and Attributes:

**Circle Class:**

Attributes:
radius: Represents the radius of the circle.

Method:
display_info: Displays information about the circle.

**Rectangle Class:**

Attributes:

length: Represents the length of the rectangle.

width: Represents the width of the rectangle.

Method:
display_info: Displays information about the rectangle.

**Triangle Class:**

Attributes:

base: Represents the base length of the triangle.

height: Represents the height of the triangle.

Method:

display_info: Displays information about the triangle.

**Sample Input:**
No direct input is provided. The input consists of creating instances of the Circle, Rectangle, and Triangle classes.

**Sample Output:**
The output should display the information about each geometric shape using the display_info method.

circle = Circle(5)

rectangle = Rectangle(4, 6)

triangle = Triangle(3, 8)

print(circle.display_info())     
print(rectangle.display_info())  
print(triangle.display_info())   


```python
class Circle:
    def __init__(self, radius):
        self.radius = radius

    def display_info(self):
        return f"Circle with radius {self.radius}"

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def display_info(self):
        return f"Rectangle with length {self.length} and width {self.width}"

class Triangle:
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def display_info(self):
        return f"Triangle with base {self.base} and height {self.height}"

circle = Circle(5)
rectangle = Rectangle(4, 6)
triangle = Triangle(3, 8)

print(circle.display_info())
print(rectangle.display_info())
print(triangle.display_info())
```

# Polymorphism - Example 1
**Problem Statement:**

Implement a simple animal sound system where different animals, such as dogs and cats, can make their characteristic sounds. Use polymorphism to allow a common interface for making sounds, with each animal providing its own implementation.

**Sample Input:**

No direct input is provided. The input consists of creating instances of the Dog and Cat classes.

**Sample Output:**

The output should display the characteristic sounds of each animal when the animal_sound function is called.

animal_sound(dog)

animal_sound(cat)

**Exaplanation**


```python
# Polymorphism - Example 1

class Animal:
    def make_sound(self):
        print("Generic animal sound")

class Dog(Animal):
    def make_sound(self):
        return "Woof! Woof!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

def animal_sound(animal):
    print(animal.make_sound())

dog = Dog()
cat = Cat()

animal_sound(dog)  # Output: Woof! Woof!
animal_sound(cat)  # Output: Meow!
```

#Polymorphism - Example 2

**Problem Statement:**

Create a class-based solution to handle addition or concatenation of two values based on their types. Implement a method **add_details** in the **CheckAdd** class, which attempts to add two parameters, based upon the value passed to them.

```python
# Polymorphism - Example 2
class CheckAdd:
    def add_details(self, x, y):
            result = x + y
            return result


adder = CheckAdd()

result1 = adder.add_details(5, 7)
result2 = adder.add_details("Hello", " World")

print("Result 1:", result1)  # Output: 12
print("Result 2:", result2)  # Output: Hello World
```

```python
#Polymorphism - using Exception Handling

class CheckAdd:
    def add_details(self, x, y):
        try:
            result = x + y
            return result
        except TypeError:
            return str(x) + str(y)


adder = CheckAdd()

result1 = adder.add_details(5, 7)
result2 = adder.add_details("Hello", " World")

print("Result 1:", result1)  # Output: 12
print("Result 2:", result2)  # Output: Hello World
```
